#1 What is hypothesis testing in statistics?
ANS: Hypothesis testing is a statistical method used to make decisions or inferences about a population parameter based on sample data. It involves formulating a hypothesis, collecting sample data, calculating a test statistic, and making a decision to accept or reject the hypothesis based on evidence.

#2 What is the null hypothesis, and how does it differ from the alternative hypothesis?
ANS:

Null hypothesis (H₀): A statement assuming no effect or no difference. It's the default assumption.

Alternative hypothesis (H₁ or Ha): A statement that contradicts the null; it represents what you're testing for — a new effect or difference.

#3 What is the significance level in hypothesis testing, and why is it important?
ANS: Significance level (α) is the probability of rejecting the null hypothesis when it is actually true (Type I error). Common values: 0.05, 0.01, 0.10. It defines how strong the evidence must be before rejecting H₀.

#4 What does a P-value represent in hypothesis testing?
ANS: The P-value is the probability of obtaining a test statistic as extreme as (or more extreme than) the one observed, assuming the null hypothesis is true.

#5 How do you interpret the P-value in hypothesis testing?
ANS:

If P-value ≤ α → reject H₀ (evidence supports Ha)

If P-value > α → fail to reject H₀ (insufficient evidence)

#6 What are Type 1 and Type 2 errors in hypothesis testing?
ANS:

Type I error (α): Rejecting H₀ when it’s actually true.

Type II error (β): Failing to reject H₀ when it’s actually false.

#7 What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
ANS:

One-tailed test: Tests for a deviation in one direction (greater than or less than).

Two-tailed test: Tests for a deviation in both directions (not equal to).

#8 What is the Z-test, and when is it used in hypothesis testing?
ANS: A Z-test compares a sample mean to a population mean when:

Population standard deviation is known.

Sample size is large (n ≥ 30).

#9 How do you calculate the Z-score, and what does it represent in hypothesis testing?
ANS:
Formula:
Z = (X̄ - μ) / (σ / √n)

Where:

X̄ = sample mean

μ = population mean

σ = population standard deviation

n = sample size

It measures how many standard errors the sample mean is away from the population mean.

#10 What is the T-distribution, and when should it be used instead of the normal distribution?
ANS: A T-distribution is used when the population standard deviation is unknown and for small sample sizes (n < 30). It has heavier tails than the normal distribution.

#11 What is the difference between a Z-test and a T-test?
ANS:

Z-test	T-test
Known population standard deviation	Unknown population standard deviation
Large sample size (n ≥ 30)	Small sample size (n < 30)
Uses normal distribution	Uses t-distribution

#12 What is the T-test, and how is it used in hypothesis testing?
ANS: A T-test evaluates whether the means of two groups are statistically different. Types include:

One-sample T-test

Independent two-sample T-test

Paired sample T-test

#13 What is the relationship between Z-test and T-test in hypothesis testing?
ANS: Both compare sample statistics to population parameters. The T-test adjusts for small sample sizes and unknown population standard deviations by using the T-distribution.

#14 What is a confidence interval, and how is it used to interpret statistical results?
ANS: A confidence interval (CI) gives a range of values within which a population parameter is likely to lie with a certain confidence level (e.g. 95%).
Example: 95% CI for mean weight: (68.5 kg, 72.3 kg)

#15 What is the margin of error, and how does it affect the confidence interval?
ANS: Margin of error (ME) is the range above and below the sample estimate.
Wider ME = wider CI (less precision)

Formula:
CI = X̄ ± ME

#16 How is Bayes' Theorem used in statistics, and what is its significance?
ANS: Bayes' Theorem updates the probability of a hypothesis based on new evidence.

Formula:
P(A|B) = [P(B|A) × P(A)] / P(B)

Used in decision-making, medical testing, and machine learning.

#17 What is the Chi-square distribution, and when is it used?
ANS: A Chi-square (χ²) distribution is used to test relationships between categorical variables and goodness of fit. It’s skewed and for non-negative values.

#18 What is the Chi-square goodness of fit test, and how is it applied?
ANS: It checks whether observed categorical data matches expected distributions.

Formula:
χ² = Σ [(O - E)² / E]

Where:

O = observed frequency

E = expected frequency

#19 What is the F-distribution, and when is it used in hypothesis testing?
ANS: An F-distribution compares two variances.
Used in:

ANOVA

F-tests

It is positively skewed.

#20 What is an ANOVA test, and what are its assumptions?
ANS: ANOVA (Analysis of Variance) tests if there are significant differences between the means of three or more groups.

Assumptions:

Independence of observations

Normality within groups

Homogeneity of variances

#21 What are the different types of ANOVA tests?
ANS:

One-way ANOVA: One independent variable

Two-way ANOVA: Two independent variables

Repeated measures ANOVA: Same subjects measured multiple times

#22 What is the F-test, and how does it relate to hypothesis testing?
ANS: An F-test compares two population variances.
In hypothesis testing:

If the F-value is larger than the critical value → reject H₀.

It’s the basis for ANOVA and regression significance testing.

```python
#1 Perform a Z-test for comparing a sample mean to a known population mean and interpret results
import numpy as np
from scipy.stats import norm

data = np.random.normal(50, 10, 30)
sample_mean = np.mean(data)
pop_mean = 50
pop_std = 10

z = (sample_mean - pop_mean) / (pop_std / np.sqrt(len(data)))
p_value = 2 * (1 - norm.cdf(abs(z)))

print("Z-score:", z)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
#4 Two-tailed Z-test and visualize decision region
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 1000)
plt.plot(x, norm.pdf(x))

plt.fill_between(x, 0, norm.pdf(x), where=(x < -1.96) | (x > 1.96), color='red', alpha=0.3)
plt.title("Two-tailed Z-test Decision Region (α=0.05)")
plt.show()
```

```python
#5 Function to calculate and visualize Type 1 & Type 2 errors
def error_regions(alpha=0.05, beta=0.2):
    x = np.linspace(-4, 4, 1000)
    plt.plot(x, norm.pdf(x), label='Null')
    plt.fill_between(x, 0, norm.pdf(x), where=(x > norm.ppf(1-alpha)), color='red', alpha=0.3, label='Type I Error')
    plt.fill_between(x, 0, norm.pdf(x, loc=2), where=(x < norm.ppf(1-alpha)), color='orange', alpha=0.3, label='Type II Error')
    plt.legend()
    plt.title("Type I and Type II Errors")
    plt.show()

error_regions()
```

```python
#9 Function to calculate confidence interval for a sample mean
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sem = stats.sem(data)
    ci = stats.t.interval(confidence, len(data)-1, loc=mean, scale=sem)
    return ci

data = np.random.normal(50, 5, 30)
print("95% Confidence Interval:", confidence_interval(data))
```

```python
#10 Calculate margin of error for given confidence level
def margin_of_error(data, confidence=0.95):
    sem = stats.sem(data)
    moe = sem * stats.t.ppf((1 + confidence) / 2, len(data)-1)
    return moe

data = np.random.normal(50, 5, 30)
print("Margin of Error:", margin_of_error(data))
```

```python
#11 Bayesian inference using Bayes' Theorem
def bayes_theorem(p_a, p_b_given_a, p_b_given_not_a):
    p_not_a = 1 - p_a
    p_b = p_b_given_a * p_a + p_b_given_not_a * p_not_a
    p_a_given_b = (p_b_given_a * p_a) / p_b
    return p_a_given_b

result = bayes_theorem(0.01, 0.9, 0.05)
print("P(A|B):", result)
```
